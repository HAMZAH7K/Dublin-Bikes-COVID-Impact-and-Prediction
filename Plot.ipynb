```python
import pandas as pd
import matplotlib.pyplot as plt

combined_df = pd.read_csv('monthly_combined.csv', parse_dates=['TIME'])

plt.figure(figsize=(10, 6))

# Plot for pre-pandemic timeline (green)
pre_pandemic_df = combined_df[combined_df['TIME'] < '2020-04']
plt.plot(pre_pandemic_df['TIME'], pre_pandemic_df['BIKE_USAGE'], label='Pre-Pandemic', color='green')

# Plot for pandemic timeline (red)
pandemic_df = combined_df[(combined_df['TIME'] >= '2020-02') & (combined_df['TIME'] < '2021-04')]
plt.plot(pandemic_df['TIME'], pandemic_df['BIKE_USAGE'], label='Pandemic', color='red')

# Plot for post-pandemic timeline (blue)
post_pandemic_df = combined_df[combined_df['TIME'] >= '2021-03']
plt.plot(post_pandemic_df['TIME'], post_pandemic_df['BIKE_USAGE'], label='Post-Pandemic', color='blue')


plt.title('City-wide Bike Usage Over Time')
plt.xlabel('Time')
plt.ylabel('Bike Usage')
plt.legend()
plt.grid(True)
plt.show()
```
